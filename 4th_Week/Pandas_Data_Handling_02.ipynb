##**Groupby 1**

- SQL groupby 명령어와 같음  
- split -> apply -> combine 과정을 거쳐 연산

```python
import numpy as np
import pandas as pd
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum()
```

```python
df.groupby(["Team", "Year"])["Points"].sum()
```

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index.index
```

```python
h_index["Devils":"Kings"]
```

```python
h_index.unstack()
```

```python
h_index.swaplevel()
```

```python
h_index.swaplevel()
```

```python
h_index.sum(level=0)
```

```python
h_index.sum(level=1)
```

##**Groupby 2**

- Groupby에 의해 split된 상태를 추출가능함

```python
grouped = df.groupby("Team")

for name,group in grouped:
    print("======================================")
    print(name)
    print("======================================")
    print(group)
    print("======================================\n")
```

```python
grouped.get_group("Devils")    # 특정 key값을 가진 그룹의 정보만 추출 가능
```

#####Groupby - aggregation : 요약된 통계정보를 추출

```python
grouped.agg(sum)
```

```python
import numpy as np
grouped.agg(np.mean)
```

```python
grouped["Points"].agg([np.sum, np.mean, np.std])
```

#####Groupby - Transformation : 해당 정보를 변환해줌

```python
df
```

```python
score = lambda x : (x)
grouped.transform(score)
```

```python
df
```

```python
score = lambda x : (x.max())
grouped.transform(score)
```

```python
score = lambda x : (x - x.mean()) / x.std()
grouped.transform(score)
```

#####Groupby - Filtration : 특정 정보를 제거 하여 보여주는 필터링 기능

```python
df
```

```python
df.groupby('Team').filter(lambda x : len(x) >= 3)
```

##**Case study**

- 시간과 데이터 종류가 정리된 통화량 데이터

```python
import dateutil

df_phone = pd.read_csv("https://raw.githubusercontent.com/TEAMLAB-Lecture/AI-python-connect/master/codes/ch_3/part-2/data/phone_data.csv")
df_phone.head()
```

```python
df_phone['data'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst = True)
df_phone.head()
```

```python
df_phone.groupby('month')['duration'].sum()
```

```python
df_phone[df_phone['item'] == 'call'].groupby('network')['duration'].sum()
```

```python
df_phone.groupby(['month', 'item'])['date'].count()
```

```python
df_phone.groupby(['month','item'])['date'].count().unstack()
```

```python
df_phone.groupby('month', as_index=False).agg({"duration":"sum"})
```

```python
df_phone.groupby(['month','item']).agg({'network_type':"count", 'duration':sum, 'date': 'first'})
```

```python
df_phone.groupby(['month','item']).agg({'network_type':"count", 'date':[min, 'first', 'unique'], 'duration':[min,max,sum]})
```

```python
grouped = df_phone.groupby('month').agg({"duration" : [min, max, np.mean]})

grouped.columns = grouped.columns.droplevel(level=0)
grouped.rename(columns={"min":"min_duration","max":"max_duration","mean": "mean_duration"})
```

##**Pivot table Crosstab**

```python
df_phone = pd.read_csv("https://raw.githubusercontent.com/TEAMLAB-Lecture/AI-python-connect/master/codes/ch_3/part-2/data/phone_data.csv")

df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.pivot_table(["duration"], index=[df_phone.month, df_phone.item], columns=df_phone.network, aggfunc="sum", fill_value=0)
```

#####Crosstab

```python
df_movie = pd.read_csv("https://raw.githubusercontent.com/TEAMLAB-Lecture/AI-python-connect/master/codes/ch_3/part-2/data/movie_rating.csv")

df_movie.head()
```

```python
pd.crosstab(index=df_movie.critic, columns=df_movie.title, values=df_movie.rating,aggfunc="first").fillna(0)
```

```python
df_movie.pivot_table(["rating"], index=df_movie.critic, columns=df_movie.title, aggfunc="sum", fill_value=0)
```

##**Merge&Concat**

#####Merge

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a, df_b, on='subject_id')
```

```python
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='left')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='right')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='inner')
```

```python
pd.merge(df_a, df_b, right_index=True, left_index=True)
```

```python
df_new = pd.concat([df_a, df_b])
df_new.reset_index()
```

```python
df_a.append(df_b)
```

```python
df_new = pd.concat([df_a, df_b], axis=1)
df_new.reset_index()
```
